```python
## CARREGAR AS BIBLIOTECAS ##
from sklearn.datasets import load_digits
import pandas as pd

## CARREGAR OS DADOS ##
dataset = load_digits()

## NOME DAS VARIAVEIS DOS DADOS ##
dataset.keys()         
```

```python
## DIMENSÃO DOS DADOS ##
dataset.data.shape
```

```python
## PLOTANDO OS DADOS PARA VISUALIZAÇÃO ## 

from matplotlib import pyplot as plt
%matplotlib inline
plt.gray()
plt.matshow(dataset.data[0].reshape(8,8))
```

```python
## DEFININDO O VETOR DE DADOS UTILIZADOS PARA UTILIZAR O PCA ##

df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
## DEFININDO DADOS DE TREINO E DE SAÍDA ##
X = df
y = dataset.target
```

```python
## APLICANDO O PCA ATRAVÉS DO SKLEARN ##
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X = pd.DataFrame(X_pca)
X.abshead()
```

#### Como podemos ver, os dados agora tem 95% da variância antiga e e tem apenas 29 dimensões ao invés de 64 como inicialmente
