# PCA - Redução de Dimensionalidade

#### Porque reduzir as dimensões dos seus dados?

        
        Um bom algortimo de Machine Learning depende de uma boa quantidade de dados para ter uma alta perfomance. Sabemos que quanto mais dados, mais eficiente é o treino e em meio a essa ideia, podemos pensar que quanto mais variaveis conseguimos extrair dos dados, melhor será a eficiência do algortimo, porém, nem sempre isso é verdade!

#### Maldição da dimensionalidade

    Imagine que você possui um dataset com milhares de características, humanamente torna-se impossível analisar a correlação entre as variáveis e a relação entre variavél e saída final. Muitas vezes também, o excesso de variáveis pode se tornar um problema por muitas dessas características dos dados serem reduntantes (por exemplo em uma tabela de dados de pessoas, existir as colunas "Idade" e "Ano de nascimento") ou até mesmo serem irrelevantes para a saída. Isso faz com que o modelo sofra um "overfitting"




    Além disso, uma grande quantidade de dados que poderiam ser desconsiderados sem graves alterações, podem acabar "pesando" o treinamento do algoritmo, fazendo com que ele se torne mais lento e as vezes, menos preciso.
    
#### PCA

    Existem várias estratégias para de redução de dimensionalidade e a mais conhecida é a PCA. Sua popularidade se deve por ser uma técnica simples, rápida e linear.
    Está técnica é baseada em criar uma nova representação dos dados, com menor dimensão e visando manter a variância entre os dados. Cada eixo de dimensão possui uma porcentagem responsável pela variância total dos dados e essa técnica visa manter esses eixos com novos valores que são calculados.
    
Vejamos a seguir um exemplo:

```python
## CARREGAR AS BIBLIOTECAS ##
from sklearn.datasets import load_digits
import pandas as pd

## CARREGAR OS DADOS ##
dataset = load_digits()

## NOME DAS VARIAVEIS DOS DADOS ##
dataset.keys()         
```

```python
## DIMENSÃO DOS DADOS ##
dataset.data.shape
```

```python
## PLOTANDO OS DADOS PARA VISUALIZAÇÃO ## 

from matplotlib import pyplot as plt
%matplotlib inline
plt.gray()
plt.matshow(dataset.data[0].reshape(8,8))
```

```python
## DEFININDO O VETOR DE DADOS UTILIZADOS PARA UTILIZAR O PCA ##

df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
## DEFININDO DADOS DE TREINO E DE SAÍDA ##
X = df
y = dataset.target
```

```python
## APLICANDO O PCA ATRAVÉS DO SKLEARN ##
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X = pd.DataFrame(X_pca)
X.abshead()
```

#### Como podemos ver, os dados agora tem 95% da variância antiga e e tem apenas 29 dimensões ao invés de 64 como inicialmente
