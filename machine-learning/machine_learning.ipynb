## Tarefas do Machine Learning
Os problemas simples resolvidos utlizando o machine learning, a depender da sua natureza, podem ser abordados utilizando-se uma das três principais estratégias desse subcampo da inteligência artificial: aprendizado supervisionado, não supervisionado e por reforço. 

Na introdução deste notebook foi citado um exemplo de classificação, essa é uma subárea do aprendizado supervisionado, a palavra supervionado nesse caso significa que disponibilizamos para o modelo exemplos daquilo que ele deve prever. Ou seja, no caso dos emails alimentamos o modelo com os textos e um indicador(label) informando se há ou não a presença de spam, com isso o modelo pode "supervisionar" as entradas para gerar suas regras de funcionamento.

Nas tarefas de aprendizado não supervisionado, por outro lado, não é possível fornecer ao modelo informações diretas da saída que deve ser fornecida. Imagine que uma empresa queira prever quais produtos tem mais chance de ser comprado por determinado clinte específico. Para isto ela dispõe do histórico de navegação do cliente em seu e-commerce, nele estão presentes compras anteriores, links acessados, tempo de permanência e setor mais visitado.
Note que, nesse caso, não há um "exemplo" de produto a ser recomendado e a relação entre o comportamento de navegação e potencial de venda de determinado produto deve ser extraída pelo modelo apenas a partir do histórico de navegação.

O aprendizado por reforço é uma tarefa semelhante ao adestramento de cães, ou seja, definimos uma métrica para "recompensar" o modelo quando ele se aproxima da resposta correta e também um modo de puni-lo quando erra. Imagine que você queira criar uma IA para zerar o jogo flappy-bird, por exemplo, nesse caso a partir da métrica de recompensas e punições a ideia é criar um pássaro que obtenha a maior puntuação possível.

## Aprendizado Supervisionado

Nesta seção resolveremos um problema de classificação para exemplificar a aplicação do machine learning em uma tarefa de aprendizado supervisionado. Neste exemplo nossa tarefa é diferenciar laranjas de toranjas de acordo com seus aspectos físicos. Informações sobre o conjunto de dados (dataset) utilizado podem ser encontradas no link abaixo 


https://www.kaggle.com/datasets/joshmcadams/oranges-vs-grapefruit

```python
# Importando bibliotecas necessárias
import pandas as pd

# Leitura do dataset
data = pd.read_csv("citrus.csv")
data.head()
```

```python
# Verificação de dados faltantes 
data.isnull().sum()
```

- Observe que os dados utilizados já estão estruturados (tabulares), não há dados faltantes nas colunas e a quantidade amostral das classes é igual. Esse dataset foi escolhido para simplifcar o entendimento e focar os esforços na modelagem em si, entretanto em um problema mais complexo os dados não seriam tão amigáveis e haveria necessidade de aplicação dos conceitos de data cleaning e feature engineering para pré-processar os dados e definir as melhores features. 

- Como as classes estão caracterizadas pelos nomes das frutas temos que binariza-las para treinar o modelo

```python
# Binarizando classes

label_mapping = {"orange": 0, "grapefruit": 1}

data["class"] = data["name"].replace(label_mapping)

data.drop(["name"], axis=1, inplace=True)

data.head()
```

# Divisão: Treino x Validação x Teste

Antes de prosseguirmos com a modelagem é importante pontuar o conceito de divisão do conjunto de dados. Quando treinamos um modelo de machine learning extraímos informações dos dados fornecidos para fazer predições, entretanto para que possamos fazer uma medição justa do desempenho do modelo temos que testa-lo com dados que não foram utilizados por ele no treinamento. Para simplificar o entendimento podemos fazer um paralelo com uma avaliação escolar, se um professor disponibiliza numa prova apenas questões abordadas nos exemplos do livro texto utilizado, essa prova não medirá corretamente o aprendizado dos alunos. Isso porque há grande chance dos alunos terem apenas memorizado as respostas. O mesmo ocorre com os modelos de machine learning, por isso separamos uma parte dos dados para testar nosso modelo. Além disso, separamos também uma parte para validação do modelo, ou seja, deixamos uma pequena parte dos dados para avaliar eventuais mudanças nas métricas da modelagem após o treinamento.

Outro ponto importante é há uma convenção sobre as nomeclaturas utilizadas na divisão dos dados, chamamos de X o conjunto de dados que alimentará o modelo (features) e de y a coluna que repesenta os exemplos de  "resposta" (target) que o modelo deve aprender a prever.

```python
# Divisão dos dados
from sklearn.model_selection import train_test_split

X, y = data.drop(["class"], axis=1), data["class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Importando modelo
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()

# Treinando modelo
model.fit(X_train, y_train)
```

```python
# Avaliando desempenho do modelo

model.score(X_test, y_test)
```

- Observe que não foi necessário decidir previamente a métrica de avaliação do modelo, isto porque utilizamos o método score do modelo KNeigbohrsClassifier fornecido pela biblioteca scikit-learn, entretando para problemas mais complexos escolher a métrica mais adequada é essencial.

```python
# Avalição do modelo utilizando dados de treino
model.score(X_train, y_train)
```

- Para fins didáticos medimos a capacidade do modelo de prever as classes ({"orange": 0 ou "grapefruit": 1) dos dados utilizados no treinamento, observe que obtivemos uma taxa maior de acerto em relação ao teste, mas o desempenho ainda ficou relativamente longe dos 100%. Esse é um bom indicativo de que o modelo de fato aprendeu a classificar as frutas e não apenas memorizou os dados de treinamento.
